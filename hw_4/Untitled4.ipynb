```python
import pandas as pd
df = pd.read_csv('train.csv')
df
```

```python
# @title Age

from matplotlib import pyplot as plt
fig, ax = plt.subplots()
ax.hist(df['Age'].dropna(), bins=30, edgecolor='black')
ax.set_title('Возраст')
ax.set_xlabel('Возраст')
ax.set_ylabel('Количество')
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots()
ax.hist(df['Survived'].dropna(), bins=2, edgecolor='black')
ax.set_title('Выживание')
ax.set_ylabel('Количество')
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots()
ax.hist(df['Pclass'].dropna(), bins=3, edgecolor='black')
ax.set_title('Класс')
ax.set_ylabel('Количество')
ax.set_xlabel('Класс')
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots()
ax.hist(df['Sex'].dropna(), bins=2, edgecolor='black')
ax.set_title('Пол')
ax.set_ylabel('Количество')
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots()
ax.hist(df['Parch'].dropna(), bins=6, edgecolor='black')
ax.set_title('Родственники')
ax.set_ylabel('Количество пассажиров')
ax.set_xlabel('Количество родственников')
```

```python
fig, ax = plt.subplots()
ax.boxplot(df['Age'].dropna(), vert=True, patch_artist=True)
ax.set_ylabel("Возраст")
ax.set_title("Распределение возраста")
```

# Проинтепретируйте полученный график
Половина пассажиров имеет возраст от 20 до 38 лет. центральная линия в коробке 28-29 лет, т.е. чуть больше половины старше, другая половина младше. нижний ус значит, что есть младенцы. Верхний ус, что есть 60+. Точки над усом это редкие значения до 80 лет.


```python
survived_counts = df['Survived'].value_counts()
pcclass_counts = df['Pclass'].value_counts().sort_index()
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].pie(survived_counts,
            labels=['Не выжил', 'Выжил'],
            autopct='%1.1f%%')
axes[0].set_title('Survived')
axes[1].pie(pclass_counts,
            labels=['1 класс', '2 класс', '3 класс'],
            autopct='%1.1f%%')
axes[1].set_title('Pclass')
```
